```python
produksi["Kabupaten"] = (
    produksi["Kabupaten"]
    .str.strip()
    .str.title()
)
```

```python
print(produksi["Kabupaten"].nunique())

sorted(produksi["Kabupaten"].unique())
```

```python
produksi.groupby("Kabupaten")["Tahun"].nunique()
```

```python
produksi.groupby("Kabupaten").size()
```

```python
produksi[
    (produksi["Tahun"]==2025)
].head(30)
```

```python
produksi["Produksi Padi"] = (
    produksi["Produksi Padi"]
    .astype(str)
    .str.replace(",", "", regex=False)   # hilangkan pemisah ribuan
    .replace("-", np.nan)                # "-" menjadi NaN
)

produksi["Produksi Padi"] = pd.to_numeric(
    produksi["Produksi Padi"],
    errors="coerce"
)
```

```python
produksi.info()
```

```python
produksi[
    (produksi.Kabupaten=="Bangkalan") &
    (produksi.Tahun==2025)
].sort_values("Bulan")
```

```python
produksi = produksi.reset_index(drop=True)
```
